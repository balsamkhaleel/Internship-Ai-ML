```python
# even or odd number
for value in [1,2,3,4]: 
    if value % 2 == 0:
        print (f"{value} is even")
    else :
        print (f"{value} is odd")
```

```python
# function to calcolatue normalize
def normalize(value , min_val , max_val):
    # Scale a value to the 0-1 range
    return (value - min_val)/(max_val - min_val)
print (normalize(75 , 0 , 100 ))     
```

Even Number 
I used a for loop to calculate the even numbers among the first 20 numbers
I used a condition (If) such that if the number is divisible by 2, it is an even number

```python
# calcolatue square number
squares = [x**2 for x in range (10)]
print (f" squares : {squares}")

# calcolatue even number
evens = [x for x in range (20) if x % 2 == 0]
print (f" evens : {evens}")
```

```python
# class represention a data record
class DataPoint:
    def __init__(self, x , y):
        self.x = x
        self.y = y
    def distance_from_origin(self):
        return (self.x**2 + self.y**2) ** 0.5
        
p = DataPoint(3,4)
print (p.distance_from_origin())
```

 I used  method to calculate the mean, the min value, and the max value

```python
def mean_min_max (numbers): 
    mean = sum (numbers)/ len(numbers)
    min_val = min(numbers)
    max_val = max(numbers)
    return  {"mean" : mean , "min": min_val ,"max": max_val}

    
number = [21 ,14 , 3 ,2 ,10 ,8 , 5] 
print (mean_min_max (number))
```

I created a class to display a specific record, and I used a function to display that record; the temperature reading was used as an example

```python
class Data_record :
    def __init__ (self , name , value):
       self.name = name 
       self.value = value
    def decribe_record (self):
        return f"{self.name} : {self.value}"

record = Data_record("temperature" , 25) 

print(record.decribe_record())     
```
