## 1. Even or Odd Numbers
Loop through a list of numbers and use an `if` condition to check whether each one is divisible by 2.

```python
# even or odd number
for value in [1, 2, 3, 4]:
    if value % 2 == 0:
        print(f"{value} is even")
    else:
        print(f"{value} is odd")
```

## 2. Normalize Function
A function that scales a value to the 0-1 range given a minimum and maximum.

```python
# function to calculate normalize
def normalize(value, min_val, max_val):
    # Scale a value to the 0-1 range
    return (value - min_val) / (max_val - min_val)


print(normalize(75, 0, 100))
```

## 3. List Comprehensions
**Even numbers:** I used a for loop to calculate the even numbers among the first 20 numbers. I used a condition (`if`) such that if the number is divisible by 2, it is an even number.

Also computing the squares of the first 10 numbers the same way.

```python
# calculate square numbers
squares = [x**2 for x in range(10)]
print(f"squares: {squares}")

# calculate even numbers
evens = [x for x in range(20) if x % 2 == 0]
print(f"evens: {evens}")
```

## 4. DataPoint Class
A class representing a data record with two coordinates and a method that computes its distance from the origin.

```python
# class representing a data record
class DataPoint:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance_from_origin(self):
        return (self.x**2 + self.y**2) ** 0.5


p = DataPoint(3, 4)
print(p.distance_from_origin())
```

## 5. Mean, Min, and Max
I used a function to calculate the mean, the min value, and the max value of a list of numbers.

```python
def mean_min_max(numbers):
    mean = sum(numbers) / len(numbers)
    min_val = min(numbers)
    max_val = max(numbers)
    return {"mean": mean, "min": min_val, "max": max_val}


number = [21, 14, 3, 2, 10, 8, 5]
print(mean_min_max(number))
```

## 6. Data Record Class
I created a class to display a specific record, and I used a method to describe that record; the temperature reading was used as an example.

```python
class DataRecord:
    def __init__(self, name, value):
        self.name = name
        self.value = value

    def describe_record(self):
        return f"{self.name} : {self.value}"


record = DataRecord("temperature", 25)
print(record.describe_record())
```
