 تم تعويض القيم المفقودة بمتوسط الحسابي للاعمار و ما تم الحذف لانه القيم المفقودة كثيرة و الداتا قليلة


```python
#التخلص من القيم المفقودة
df["Age"].fillna(df["Age"].mean(),inplace = True)
df["Age"].isnull().sum()
```

```python
# drop Cabin
df=df.drop(columns=["Cabin"])
print(df.columns)
```
