# Day 3 — NumPy Array Manipulation

## 1. Array Creation
Create arrays using several of NumPy's built-in constructors.

```python
import numpy as np

a = np.array([1, 2, 3, 4])
z = np.zeros((2, 3))
o = np.ones(5)
r = np.arange(0, 10, 2)
l = np.linspace(0, 1, 5)
rnd = np.random.rand(3, 3)

print(a)
print(z)
print(o)
print(r)
print(l)
print(rnd)
```

## 2. Indexing & Slicing (2D)
Access elements, rows, and columns of a 2D array using `array[row, column]` syntax.

```python
m = np.array([[1, 2, 3], [4, 5, 6]])
print(m.shape)
print(m[0, 2])
print(m[:, 1])
print(m[1, :])
```

## 3. Vectorized Operations
Operations apply to the whole array at once, with no explicit loop.

```python
a = np.array([1, 2, 3, 4])
print(a * 2)
print(a + a)
print(a.mean())
print(a[a > 2])
```

## 4. Broadcasting
Add a 1D row array to every row of a 2D matrix without writing a loop.

```python
matrix = np.array([[1, 2, 3], [4, 5, 6]])
row = np.array([10, 20, 30])
print(matrix + row)
```

## 5. Hands-On Lab: NumPy Array Manipulation

### Step 1 — Shape and Dtype
Build a 4x4 matrix and inspect its `shape` and `dtype`.

```python
matrix = np.array([[5, 6, 8, 1], [10, 4, 2, 11], [12, 13, 14, 15], [4, 6, 10, 8]])

print(f"shape: {matrix.shape}")
print(f"data type: {matrix.dtype}")
```

### Step 2 — Slicing
Extract the third column and the last row.

```python
print(matrix[:, 2])
print(matrix[3, :])
```

### Step 3 — Boolean Masking
Select all values greater than the matrix's mean.

```python
matrix_mean = matrix.mean()
print(f"matrix mean = {matrix_mean}")
print(matrix[matrix > matrix_mean])
```

### Step 4 — Broadcasting
Add a 1D array to every row of a 2D array.

```python
array1 = np.array([[3, 4, 5, 6], [1, 10, 5, 9]])
array2 = np.array([6, 8, 2, 4])
print(array1 + array2)
```
